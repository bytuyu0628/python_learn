```python
def get_filename(path,filetype):
    name =[]
    final_name = []
    for root,dirs,files in os.walk(path):
        for i in files:
            if filetype in i:
                wholePath = root +'/'+ i.replace(filetype,'')
                name.append(wholePath.replace(path, ''))#生成不带filetype后缀的文件名组成的列表
    final_name = [item +fileType for item in name]#生成filetype后缀的文件名组成的列表
    return final_name#输出由有filetype后缀的文件名组成的列表
```

判断字符串中是否包含中文

```python
import re
def containCN(word):
    zhmodel = re.compile(u'[\u4e00-\u9fa5]')     
    match = zhmodel.search(word)
    return match
```

```python
# 项目地址
projectDir = '/Users/taodi/Documents/projects/ct_ctw_android/app/src/main/java/com/tts/trip'
fileType = '.java'
outputXmlFilename = 'string.xml'
fileList = []
exist = os.path.exists(projectDir)
count = 1
if exist:
    print(projectDir, '项目目录存在');
    fileList = get_filename(projectDir, fileType)
    print('找到%d份java文件' % len(fileList))
    fp = open(outputXmlFilename,'w+')
    for file in fileList:
        print('开始分析', file.replace(projectDir, ' '))
        print('\033[1;33m>>>>>>>\033[0m')
        totalString = '字符串：'
        for line in open(projectDir + file):
            words = line.split('"')
            if len(words) >= 3:
                for index in range(len(words)):
                    #非单数分割段的，不是字符串内容
                    if index%2 != 1:
                        continue
                    #去除长度小于4的字符串
                    if len(words[index]) < 4: 
                        continue
                    #前一个分割段
                    preWord = words[index-1]
                    #去除声明的字符串常量
                    if '=' in preWord:
                        continue
                    #只保留中文
                    if not containCN(words[index]):
                        continue
                    totalString += '"'+words[index] +'" '
                    fp.write('<string name="%s">%s</string>\n' % ('str'+str(count), words[index]))
                    count = count + 1
        if(totalString != '字符串：'):
            print(totalString)
        print('\033[1;31;43mありがとうpython--------------------------------つづく\033[0m')
    fp.close()
```
